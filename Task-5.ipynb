## Exploratory Data Analysis - Sports

### First, We need to import the necessary libraries and load the dataset into our  Python environment

```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the dataset into a pandas dataframe

ipl_data = pd.read_csv('ipl matches.csv')
```

```python
# Displaying the csv dataset

display(ipl_data)
```

###  Let's take a quick look at the dataset 

```python
# Displaying first 5 rows of the dataset

ipl_data.head()
```

### We need to perform some data cleaning and preprocessing steps.

```python
# Checking for missing values

ipl_data.isnull().sum
```

```python
# Dropping unnecessary columns

ipl_data.drop(['id', 'umpire1', 'umpire2', 'umpire3'], axis=1, inplace=True)
```

```python
# Displaying the dropping unnecessary columns

display(ipl_data.drop)
```

```python
# Replacing team names with their abbreviations


ipl_data.replace({'Mumbai Indians': 'MI', 'Kolkata Knight Riders': 'KKR',
                  'Royal Challengers Bangalore': 'RCB', 'Deccan Chargers': 'DCH',
                  'Chennai Super Kings': 'CSK', 'Rajasthan Royals': 'RR',
                  'Delhi Daredevils': 'DD', 'Kings XI Punjab': 'KXIP',
                  'Sunrisers Hyderabad': 'SRH', 'Rising Pune Supergiant': 'RPS',
                  'Gujarat Lions': 'GL', 'Rising Pune Supergiants': 'RPS'}, inplace=True)
```

```python
# Displaying team names with their abbreviations

display(ipl_data.replace)
```

```python
# Renaming column names for consistency

ipl_data.rename(columns={'win_by_runs': 'Win_by_Runs', 'win_by_wickets': 'Win_by_Wickets',
                         'player_of_match': 'Player_of_Match', 'venue': 'Venue',
                         'winner': 'Winner', 'result': 'Result', 'dl_applied': 'DL_Applied',
                         'season': 'Season', 'city': 'City', 'date': 'Date',
                         'toss_winner': 'Toss_Winner', 'toss_decision': 'Toss_Decision',
                         'team1': 'Team1', 'team2': 'Team2'}, inplace=True)
```

```python
#  Displaying the Renaming column names for consistency

display(ipl_data.rename)
```




### Let's start with our analysis.


## Most Successful Teams

### Let's first determine the most successful teams in IPL history by looking at the number of titles won.

```python
# Grouping the data by the winning team and season and counting the number of wins

team_wins = ipl_data.groupby(['Winner', 'Season']).size().reset_index(name='Wins')
```

```python
# Displaying the Grouping the data by the winning team and season and counting the number of wins

display(team_wins)
```

```python
# Grouping the data by winning team and calculating the total number of wins

team_wins = team_wins.groupby(['Winner'])['Wins'].sum().reset_index()
```

```python
# Displaying Grouping the data by winning team and calculating the total number of wins

display(team_wins)
```

```python
# Shorting the teams based on their total number of wins

team_wins = team_wins.sort_values(by=['Wins'], ascending=False)
```

```python
# Shorting the teams based on their total number of wins

display(team_wins)
```

```python
# Plotting the bar chart

plt.figure(figsize=(12,6))
sns.barplot(x='Winner', y='Wins', data=team_wins, palette='Paired')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Most Successful Teams in IPL History')
plt.show()
```
