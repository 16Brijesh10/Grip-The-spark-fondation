## Now, move on to more advanced EDA techniques to answer the business questions. The  busuness problems we can derive from exploring the data include:

### 1) Which product categories are the most profitable?

### 2) Which regions have the highest sales and profit margins?

### 3) Which shipping model is the most commonly used and has the highest profit margin?

### 4) What is the correlation between variables like sales, profit, and quantity sold?

### 5) Which states haves the highest and lowest sales and profit margins?



## To answer these question , we can make use of various visualization techniques such as scatter plots , bar charts, and heatmaps. 
    

```python
#Bar chart showing for total sales by product category

sns.barplot(x="Category", y="Sales", data=df, estimator=sum)
plt.title("Total sales by product category")
plt.show()
```

```python
#Heatmap showing correlaation between sales , profit and quantity

sns.heatmap(df[["Sales", "Profit", "Quantity"]].corr(), annot=True)
plt.title("Correlation between Sales, Profit and Quantity")
plt.show()
```

```python
#Bar chart showing profit margin by shipping mode

sns.barplot(x="Ship Mode", y="Profit", data=df)
plt.title("Profit Margin by Shipping Mode")
plt.show()
```

```python
#scatter plot showing sales and profit by state

sns.scatterplot(x="Sales", y="Profit", hue="State", data=df)
plt.title("Sales and Profit by State")
plt.show()
```

### Based on the analysis and visualizations, we can identify the weak areas and opportunities to increase profit.
